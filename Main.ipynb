```python
GP = mat[mat["school"]=="GP"]
mediagp = GP["G3"].mean()
MS = mat[mat["school"]=="MS"]
mediams = MS["G3"].mean()
cont1 = 0
cont2 = 0
for nota in GP["G3"]:
    if nota>=9.5:
        cont1=cont1+1
for nota in MS["G3"]:
    if nota>=9.5:
        cont2=cont2+1
porc1=(cont1*100)/len(GP["G3"])        
porc2=(cont2*100)/len(MS["G3"]) 
print "La nota media en los estudiantes de la ecuela GP es %f y el porcentaje de aprobados %f , y en la escuela MS es %f el porcentaje de aprobados es %f"%(mediagp,porc1,mediams,porc2)
```

```python
hg = mat[mat["higher"]=="yes"]
nhg = mat[mat["higher"]=="no"]
media1 = hg["G3"].mean()
media2 = nhg["G3"].mean()
print "La nota media de los estudiantes que tienen interes en realizar estudios superiores es %f , y en los estudiantes que no tienen interes es %f"%(media1,media2)
```
